## Importing data

```python
data = data_load('Unnamed: 0','BRCA1P1','BRCA1',200)
```

The data parameter must be a dictionary containing:

Source domain: 'z' (n x p), 'X' (n,), 'Y' (n,)

External domain: 'zE' (n_E x p), 'XE' (n_E,), 'YE' (n_E,)

Unlabeled domain: 'zU' and 'XU' as lists of arrays with matching structures. Each list element corresponds to a separate unlabeled data source. For example, 'zU' is a list of arrays each with dimensions (n_U_i x p), and 'XU' is a list of arrays each with length n_U_i. The list structures of 'zU' and 'XU' must match exactly (same number of elements), and corresponding elements within the lists must have consistent sample sizes.

Where n is the number of source samples, n_E is the number of external samples, p is the latent dimension, and n_U_i represents the number of samples in the i-th unlabeled source.
