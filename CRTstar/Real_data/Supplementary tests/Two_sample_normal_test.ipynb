```python
import numpy as np
import scipy.stats as stats
import warnings
import sys
import os

current_dir = os.getcwd()
sys.path.append(os.path.dirname(current_dir))
sys.path.append(os.path.dirname(os.path.dirname(current_dir)))
from read_data_package.utils import *
from MYCRT_package.utils import *
from config import *

warnings.filterwarnings("ignore")
```

```python
def T_statistics(X, Y, z ,lam_const, coef = None):
    n , _ = z.shape
    coefy , _ = fit_lasso_model(z,Y,lam_const)
    dy = Y-np.dot(z,coefy)
    coef , lam_const1 = fit_lasso_model(z,X,lam_const)
    dx = X-np.dot(z,coef)
    Ri = dx*dy
    tau_N = np.sqrt(n)*np.mean(Ri)
    tau_D = np.sqrt(np.mean(Ri**2)-np.mean(Ri)**2)
    return tau_N/tau_D
```

```python
data = data_load('Unnamed: 0','BRCA1P1','BRCA1',200)
z, zE = data['z'], data['zE']
X, XE = data['X'], data['XE']
Y, YE = data['Y'], data['YE']
```

```python
T_in = T_statistics(X, Y, z ,lam_const)
T_out = T_statistics(XE, YE, zE ,lam_const)
T_in, T_out
```

```python
p_value_in = 1 - stats.norm.cdf(T_in)
print(f"T_in = {T_in:.6f}")
print(f"One-sided p-value = {p_value_in:.10f}")
print(f"Conclusion: {'Reject H0' if p_value_in < 0.05 else 'Fail to reject H0'}")
if p_value_in < 0.05:
    print("     Significant evidence that mean ≥ 0")
else:
    print("     Insufficient evidence that mean ≥ 0")
print()

p_value_out = 1 - stats.norm.cdf(T_out)
print(f"T_out = {T_out:.6f}")
print(f"One-sided p-value = {p_value_out:.6f}")
print(f"Conclusion: {'Reject H0' if p_value_out < 0.05 else 'Fail to reject H0'}")
if p_value_out < 0.05:
    print("     Significant evidence that mean ≥ 0")
else:
    print("     Insufficient evidence that mean ≥ 0")
```

```python
print("=== Two-sided test for standard normal distribution (H0: μ = 0 vs H1: μ ≠ 0) ===\n")

# Test for T_in
p_value_in = 2 * (1 - stats.norm.cdf(abs(T_in)))
print(f"T_in = {T_in:.6f}")
print(f"|T_in| = {abs(T_in):.6f}")
print(f"Z-statistic = {T_in:.6f}")
print(f"Two-sided p-value = {p_value_in:.10f}")
print(f"Conclusion: {'Reject H0' if p_value_in < 0.05 else 'Fail to reject H0'}")
if p_value_in < 0.05:
    print("     Significant evidence that mean ≠ 0")
else:
    print("     Insufficient evidence that mean ≠ 0")
print()

# Test for T_out
p_value_out = 2 * (1 - stats.norm.cdf(abs(T_out)))
print(f"T_out = {T_out:.6f}")
print(f"|T_out| = {abs(T_out):.6f}")
print(f"Z-statistic = {T_out:.6f}")
print(f"Two-sided p-value = {p_value_out:.6f}")
print(f"Conclusion: {'Reject H0' if p_value_out < 0.05 else 'Fail to reject H0'}")
if p_value_out < 0.05:
    print("     Significant evidence that mean ≠ 0")
else:
    print("     Insufficient evidence that mean ≠ 0")
```
